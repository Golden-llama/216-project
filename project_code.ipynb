# What is the relationship between social factors and the mental health of university students, and how does it vary across different mental health conditions?

## Part 1: Data Normalization
This project uses two datasets from Kaggle: "University Student Mental Health Indicators" and "Student Stress Factors: A Comprehensive Analysis". As expected, the data across the datasets is measured using different scales but, even within each dataset there are different scales across columns. For example, in "University Student Mental Health Indicators", the column `PHQ9` (depression) has a range of 0–27 to align with the actual PHQ9 test developped by the American Psychiatric Association. However, many columns developped by the researchers like `FamilySupport` are from 0–1. As is, comparing values across columns is meaningless.

Normalization will rescale every column from 0–1, so values mean the same thing everywhere.

### Importing Data

````python
import pandas as pd
import numpy as np

university_student_mental_health_indicators = pd.read_csv('data.csv')
student_stress_factors = pd.read_csv('StressLevelDataset.csv')
````

### Inspect Datasets
The minimum and maximum value in each dataset provide the range for normalization

````python
print(university_student_mental_health_indicators.describe().loc[['min','max']]) 
print(student_stress_factors.describe().loc[['min','max']])
````

### Normalize Both Datasets
Data is normalized the following formula:

$normalized x = \frac{x - min(x)}{max(x) - min(x)}$

````python
def normalize(df):
    copied_df = df.copy()  # .copy() prevents changes from affecting the original
    cols = copied_df.columns
    for col in cols:
        col_min = copied_df[col].min()
        col_max = copied_df[col].max()

        copied_df[col] = (copied_df[col] - col_min) / (col_max - col_min)
        return copied_df

normalized_university_student_mental_health_indicators = normalize(university_student_mental_health_indicators)
normalized_student_stress_factors = normalize(student_stress_factors)

print(normalized_university_student_mental_health_indicators)
print(normalized_student_stress_factors)
````

## Why These Two Methods

**Step 1 — Correlation:** Before running any formal test, we check correlations between all social factors and all outcomes at once. This tells us which social factor → outcome pair has the strongest relationship, so we are choosing our t-test target based on evidence rather than guessing. Running a test on a randomly chosen pair without checking correlations first is bad statistical practice.

**Step 2 — T-test:** A t-test asks whether two groups have meaningfully different means. We split students into 'high' and 'low' groups on a social factor (split at the median — explained below), then test whether their mental health outcome scores differ significantly. If the p-value is below 0.05, we can say the difference is statistically significant — it is unlikely to have happened by chance.

We chose a t-test over chi-squared because our outcomes (PHQ9, anxiety_level) are continuous numerical scores, not categories. Chi-squared is for two categorical variables. We chose t-test over full regression because regression requires more assumptions and interpretation — the t-test gives a clear yes/no answer to our most direct question.

````python
# .groupby() splits the DataFrame into groups based on a column's values
# .mean() then calculates the average of each social factor within each group
# .reset_index() turns the group labels back into a regular column
summary_data = (
    data_norm.groupby('MentalHealthStatus')[data_social_factors]
    .mean()
    .round(3)
    .reset_index()
)

# Replace 0/1 with readable labels
summary_data['MentalHealthStatus'] = summary_data['MentalHealthStatus'].map(
    {0.0: 'No Diagnosis', 1.0: 'Has Diagnosis'}
)

print('=== Mean Social Factor Scores by Mental Health Status (data.csv) ===')
print(summary_data.to_string(index=False))
````

### Step 2 — T-Test

**Why split at the median:** We divide students into 'high' and 'low' groups on the social factor using the median as the cutoff. The median always splits the data exactly in half regardless of scale, making both groups equal in size. This is fairer than choosing an arbitrary number like 5.

**What the p-value means:** If `p < 0.05`, there is less than a 5% chance the difference between groups happened randomly. This is the standard threshold in social science research.

**FREEDOM:** Change `social_col` and `outcome_col` to the pair that had the strongest correlation in the table above. Write your reasoning in the markdown cell below.

````python
def run_ttest(df, social_col, outcome_col, dataset_label):
    # Split at the median — half of students are in each group
    median_val  = df[social_col].median()
    high_group  = df[df[social_col] >= median_val][outcome_col].dropna()
    low_group   = df[df[social_col] <  median_val][outcome_col].dropna()

    # stats.ttest_ind compares the means of two independent groups
    t, p = stats.ttest_ind(high_group, low_group)

    print(f'\n{"="*52}')
    print(f'T-TEST [{dataset_label}]')
    print(f'  Social factor : {social_col}')
    print(f'  Outcome       : {outcome_col}')
    print(f'{"="*52}')
    print(f'  High {social_col} group mean : {high_group.mean():.4f}  (n={len(high_group)})')
    print(f'  Low  {social_col} group mean : {low_group.mean():.4f}  (n={len(low_group)})')
    print(f'  T-statistic : {t:.4f}')
    print(f'  P-value     : {p:.4f}')
    print(f'  {"SIGNIFICANT (p < 0.05)" if p < 0.05 else "Not significant"}')

    results.append({
        'dataset':       dataset_label,
        'social_factor': social_col,
        'outcome':       outcome_col,
        'high_mean':     round(high_group.mean(), 4),
        'low_mean':      round(low_group.mean(),  4),
        't_stat':        round(t, 4),
        'p_value':       round(p, 4),
        'significant':   p < 0.05
    })
````

````python
# T-TEST 1: data.csv
# FREEDOM: change social_col and outcome_col based on the correlation table above
# Pick the pair with the strongest correlation and explain why below
run_ttest(
    df=data_norm,
    social_col='FamilySupport',   # FREEDOM: change to strongest correlator
    outcome_col='PHQ9',           # FREEDOM: change to the outcome it correlates with
    dataset_label='data.csv'
)
````

````python
# T-TEST 2: StressLevelDataset
# FREEDOM: same — choose based on the stress correlation table
run_ttest(
    df=stress_norm,
    social_col='social_support',  # FREEDOM: change to strongest correlator
    outcome_col='anxiety_level',  # FREEDOM: change to the outcome it correlates with
    dataset_label='StressLevelDataset'
)
````
