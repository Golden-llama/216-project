## Part 1: Data Normalization
This project uses two datasets from Kaggle: "University Student Mental Health Indicators" and "Student Stress Factors: A Comprehensive Analysis". As expected, the data across the datasets is measured using different scales but, even within each dataset there are different scales across columns. For example, in "University Student Mental Health Indicators", the column `PHQ9` (depression) has a range of 0–27 to align with the actual PHQ9 test developped by the American Psychiatric Association. However, many columns developped by the researchers like `FamilySupport` are from 0–1. As is, comparing values across columns is meaningless.

Normalization will rescale every column from 0–1, so values mean the same thing everywhere.

### Importing Data

```python
import pandas as pd
import numpy as np

university_student_mental_health_indicators = pd.read_csv('data/data.csv')
student_stress_factors = pd.read_csv('data/StressLevelDataset.csv')
```

### Inspect Datasets
The minimum and maximum value in each dataset provide the range for normalization

```python
print(university_student_mental_health_indicators.describe().loc[['min','max']]) 
print(student_stress_factors.describe().loc[['min','max']])
```

### Normalize Both Datasets
Data is normalized the following formula:

$normalized x = \frac{x - min(x)}{max(x) - min(x)}$

```python
def normalize(df):
    copied_df = df.copy()  # .copy() prevents changes from affecting the original
    cols = copied_df.columns
    for col in cols:
        col_min = copied_df[col].min()
        col_max = copied_df[col].max()

        copied_df[col] = (copied_df[col] - col_min) / (col_max - col_min)
    return copied_df

normalized_university_student_mental_health_indicators = normalize(university_student_mental_health_indicators)
normalized_student_stress_factors = normalize(student_stress_factors)

print(normalized_university_student_mental_health_indicators)
print(normalized_student_stress_factors)
```
