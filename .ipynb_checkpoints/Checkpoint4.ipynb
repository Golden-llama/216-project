## Why GroupBy Here
Person 3's charts show individual student-level patterns. Person 4's tests confirm whether those patterns are statistically real. Person 5's job is to **summarize at the group level** — grouping students by their mental health condition and asking: do students in different condition groups experience social factors differently? This is the most direct answer to our research question's second half: 'how does it vary across different mental health conditions?'

GroupBy is the right tool because it lets us collapse thousands of rows into a few meaningful group summaries, making the pattern readable and reportable.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load all needed files
data_norm   = pd.read_csv('data_normalized.csv')
stress_norm = pd.read_csv('stress_normalized.csv')
stats_df    = pd.read_csv('stats_results.csv')

# Paste Person 2's variable lists here exactly
data_social_factors   = ['FamilySupport', 'SocialActivity', 'PeerRelationship', 'OnlineStress', 'FinancialStress']
data_mh_outcomes      = ['PHQ9', 'GAD7', 'MentalHealthStatus']
stress_social_factors = ['social_support', 'peer_pressure', 'extracurricular_activities', 'teacher_student_relationship']
stress_mh_outcomes    = ['anxiety_level', 'depression', 'stress_level']

sns.set_theme(style='whitegrid')
```

### GroupBy Summary — data.csv

We group by `MentalHealthStatus` (0 = no diagnosis, 1 = has diagnosis) and calculate the mean of each social factor per group. This tells us whether students with a mental health diagnosis experience social factors differently — lower family support, worse peer relationships, higher online stress — compared to those without a diagnosis.

**FREEDOM:** You may instead group by a binned version of `PHQ9` using `pd.cut()` to create low/medium/high depression groups. This would give three groups instead of two and might show a more gradual trend. If you do this, explain why that grouping is more informative than the binary diagnosis split.

```python
# .groupby() splits the DataFrame into groups based on a column's values
# .mean() then calculates the average of each social factor within each group
# .reset_index() turns the group labels back into a regular column
summary_data = (
    data_norm.groupby('MentalHealthStatus')[data_social_factors]
    .mean()
    .round(3)
    .reset_index()
)

# Replace 0/1 with readable labels
summary_data['MentalHealthStatus'] = summary_data['MentalHealthStatus'].map(
    {0.0: 'No Diagnosis', 1.0: 'Has Diagnosis'}
)

print('=== Mean Social Factor Scores by Mental Health Status (data.csv) ===')
print(summary_data.to_string(index=False))
```

### GroupBy Summary — StressLevelDataset

We group by `stress_level` (three categories: low/medium/high) and calculate mean social factor scores per group. This shows how social factors shift across a spectrum of mental health severity — not just diagnosed vs not, but across a gradient. Comparing these two groupings (binary vs three-level) across the two datasets strengthens the conclusion.

**FREEDOM:** You may also group by `peer_pressure` binned into low/medium/high using `pd.cut()` if you want to see how the *social factor itself* varies across groups, rather than grouping by outcome. Write your reasoning for whichever approach you choose.

```python
summary_stress = (
    stress_norm.groupby('stress_level')[stress_social_factors]
    .mean()
    .round(3)
    .reset_index()
)
summary_stress['stress_level'] = summary_stress['stress_level'].map(
    {0.0: 'Low Stress', 0.5: 'Medium Stress', 1.0: 'High Stress'}
)

print('=== Mean Social Factor Scores by Stress Level (StressLevelDataset) ===')
print(summary_stress.to_string(index=False))
```

### Statistical Results Summary

```python
# Display Person 4's test results in the final notebook
print('=== T-Test Results (from Person 4) ===')
print(stats_df.to_string(index=False))
```

### Final Chart

**Why this chart:** The GroupBy summaries above are tables — clear but not visual. This chart converts the stress-level summary into a grouped bar chart so the reader can immediately see which social factors change most across stress level categories. Grouped bars allow direct comparison of multiple social factors side by side, which tells a richer story than showing one factor at a time.

**FREEDOM:** You may use a different chart type here. A line plot with `kind='point'` would show trends more clearly if the x-axis has a natural order (which stress levels do). Write one sentence justifying whichever type you choose.

```python
# pd.melt reshapes the wide summary table into long format
# Wide:  one row per stress level, one column per social factor
# Long:  one row per stress level per social factor
# seaborn needs long format to use hue for coloring by social factor
plot_melted = pd.melt(
    summary_stress,
    id_vars='stress_level',
    value_vars=stress_social_factors,  # FREEDOM: reduce this list to the 2-3 most important factors
    var_name='social_factor',
    value_name='mean_score'
)

# Cleaner labels for the legend
label_map = {
    'social_support':               'Social Support',
    'peer_pressure':                'Peer Pressure',
    'extracurricular_activities':   'Extracurricular',
    'teacher_student_relationship': 'Teacher Relationship'
}
plot_melted['social_factor'] = plot_melted['social_factor'].map(label_map)

# FREEDOM: change kind='bar' to kind='point' for a line-style chart
sns.catplot(
    x='social_factor',
    y='mean_score',
    hue='stress_level',
    kind='bar',
    data=plot_melted,
    hue_order=['Low Stress', 'Medium Stress', 'High Stress'],
    height=5, aspect=2.0,
    palette='Set2'  # FREEDOM: change color palette
)
plt.title('Mean Social Factor Scores Across Stress Level Categories', fontsize=13)
plt.xlabel('Social Factor')
plt.ylabel('Mean Normalized Score')
plt.xticks(rotation=15)
plt.legend(title='Stress Level', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.savefig('chart_final.png', dpi=150)
plt.show()
```
