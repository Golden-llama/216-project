## Why These Chart Types
Our proposal planned scatterplots and boxplots specifically:
- **Scatterplots** show how two numerical variables relate to each other across all individual students. They are the right choice when both variables are continuous (not categories) and you want to see the direction and strength of the relationship.
- **Boxplots** show the *distribution* of a mental health score across groups. They are the right choice when one variable is categorical (like stress level: low/medium/high) and you want to compare spread, median, and outliers between groups. Our proposal mentions using boxplots to find how prevalent anxiety and stress are — boxplots show this better than a bar chart because they reveal the full spread, not just the mean.

We chose these over histograms (which only show one variable at a time) and bar charts (which hide the distribution by only showing the mean).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Person 1's normalized files
data_norm   = pd.read_csv('data_normalized.csv')
stress_norm = pd.read_csv('stress_normalized.csv')

# Paste Person 2's variable lists here exactly
data_social_factors   = ['FamilySupport', 'SocialActivity', 'PeerRelationship', 'OnlineStress', 'FinancialStress']
data_mh_outcomes      = ['PHQ9', 'GAD7', 'MentalHealthStatus']
stress_social_factors = ['social_support', 'peer_pressure', 'extracurricular_activities', 'teacher_student_relationship']
stress_mh_outcomes    = ['anxiety_level', 'depression', 'stress_level']

sns.set_theme(style='whitegrid')
```

### Chart 1 — Scatterplot (data.csv)

**Why a scatterplot here:** Both `FamilySupport` and `PHQ9` are continuous numerical variables — every student has a unique value on a sliding scale. A scatterplot lets us see each individual student as a dot, showing the overall trend across the whole sample rather than summarizing it. We add a trend line (`lmplot`) so the direction of the relationship is immediately visible.

**Why FamilySupport vs PHQ9:** Family support is one of the most direct social factors named in our proposal. PHQ9 is a clinically validated depression measure, making it the most reliable outcome in this dataset.

**FREEDOM:** You may change the social factor (x-axis) or outcome (y-axis) to any column from Person 2's lists. If you do, update the written justification below to explain why your pair is a meaningful research question.

```python
# FREEDOM: change x and y to any social factor and outcome from Person 2's lists
# If you change them, update the markdown cell above with your reasoning
x_col = 'FamilySupport'   # social factor (x-axis)
y_col = 'PHQ9'            # mental health outcome (y-axis)

# lmplot draws both a scatterplot and a trend line
# hue='MentalHealthStatus' colors dots by diagnosis — lets us see if
# the relationship looks different for diagnosed vs non-diagnosed students,
# which directly addresses 'varies across conditions' in the research question
sns.lmplot(
    x=x_col,
    y=y_col,
    hue='MentalHealthStatus',
    data=data_norm,
    scatter_kws={'alpha': 0.3, 's': 15},  # alpha = transparency so overlap is visible
    height=5, aspect=1.6,
    legend=False
)
plt.title(f'{x_col} vs {y_col} by Mental Health Status', fontsize=13)
plt.xlabel(f'{x_col} (normalized)')
plt.ylabel(f'{y_col} (normalized)')
plt.legend(title='Mental Health Status', labels=['No Diagnosis', 'Has Diagnosis'], loc='upper right')
plt.savefig('chart1.png', dpi=150)
plt.show()
```

### Chart 2 — Boxplot (StressLevelDataset)

**Why a boxplot here:** `stress_level` divides students into three categories: low, medium, and high. When one variable is categorical like this, a boxplot is the right choice — it shows the full distribution of the outcome (median, spread, outliers) within each group. A bar chart would only show the mean, hiding whether scores are tightly clustered or widely spread within each stress level.

**Why stress_level as the grouping variable:** It lets us directly compare anxiety and depression scores across mental health severity levels, which addresses the 'how does it vary across different mental health conditions' part of the research question.

**FREEDOM:** You may choose to show only `anxiety_level`, only `depression`, or both side by side (using `pd.melt` as shown below). You may also choose a different social factor as the grouping variable — for example, grouping by high vs low `peer_pressure` instead of stress_level. Write your reasoning.

```python
# Map stress_level numbers to readable labels
# stress_level is 0, 1, 2 — we replace with Low/Medium/High for the chart
stress_norm['stress_label'] = stress_norm['stress_level'].map(
    {0.0: 'Low Stress', 0.5: 'Medium Stress', 1.0: 'High Stress'}
)

# pd.melt reshapes the data so both outcomes appear side by side
# Before melt: one row per student, anxiety_level and depression as separate columns
# After melt:  two rows per student — one labeled 'Anxiety', one labeled 'Depression'
# This lets seaborn color bars by condition using hue
# FREEDOM: change value_vars to only ['anxiety_level'] or only ['depression']
# if you only want to show one condition
plot_stress = stress_norm.copy()
plot_melted = pd.melt(
    plot_stress,
    id_vars=['stress_label'],
    value_vars=['anxiety_level', 'depression'],  # FREEDOM: change this list
    var_name='condition',
    value_name='score'
)
plot_melted['condition'] = plot_melted['condition'].map(
    {'anxiety_level': 'Anxiety', 'depression': 'Depression'}
)

sns.catplot(
    x='stress_label',
    y='score',
    hue='condition',
    kind='box',
    data=plot_melted,
    order=['Low Stress', 'Medium Stress', 'High Stress'],
    height=5, aspect=1.8,
    palette={'Anxiety': 'steelblue', 'Depression': 'salmon'}  # FREEDOM: change palette
)
plt.title('Anxiety and Depression Scores Across Stress Level Categories', fontsize=13)
plt.xlabel('Stress Level Category')
plt.ylabel('Normalized Score')
plt.savefig('chart2.png', dpi=150)
plt.show()
```
